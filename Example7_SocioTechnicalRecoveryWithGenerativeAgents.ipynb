# Example 7: Socio-technical regional recovery simulation with generative household agents

Example 7 demonstrates a **socio-technical regional recovery simulation** that couples physical infrastructure recovery with human behavioral dynamics. The community modeled is a subset of Alameda Island, CA, subjected to an earthquake scenario.

**Technical systems** include residential buildings, which provide shelter, and the water supply system, which distributes potable water to households and supports post-disaster repair activities. Building and pipe damage are derived from R2D damage assessments.

**Social systems** are represented as households modeled as **generative agents** whose post-disaster displacement decisions — whether to stay at home, stay with a friend, or leave town — are simulated using a large language model (LLM). Each agent is initialized with socioeconomic characteristics (tenure, income, household composition, immigration status) and makes decisions based on its current situation, including building damage, access to infrastructure services, and the behavior of neighboring households.

The **time step is one week**, reflecting the temporal scale at which households are assumed to reassess their living situation during recovery simulation.

This example shows how **pyrecodes** integrates with the **OpenAI API** to simulate household decision-making as part of a regional recovery simulation. Household decisions feed back into the simulation by affecting the demand for shelter and potable water, coupling human mobility with infrastructure recovery dynamics.

Please refer to the **pyrecodes** [Example 7 page](https://nikolablagojevic.github.io/pyrecodes/html/usage/examples/example_7.html) for further details.

```python
from pyrecodes import main

system = main.run(folder_name='Example 7', main_file_name='SmallAlameda_Main.json')
system.calculate_resilience()
system.save_as_pickle('baseline_results.pkl')
```

```python
from pyrecodes.plotter.concrete_plotter import ConcretePlotter

plotter = ConcretePlotter()
plotter.plot_household_location_over_time(system.components, x_axis_label='Weeks after Earthquake')
plotter.plot_met_demand_over_time(system.components, 'Shelter')
plotter.plot_supply_demand_dynamics(system, ['PotableWater'], units=['m3/week'])
```

```python
from pyrecodes.geovisualizer.household_geovisualizer import HouseholdGeoVisualizer

residential_buildings = [component for component in system.components if type(component).__name__ == 'R2DBuildingWithHouseholds']
geo_visualizer = HouseholdGeoVisualizer(residential_buildings)

step = 5 # create a figure every 5 time steps
time_steps = range(system.START_TIME_STEP, system.time_step + 1, step)
for time_step in time_steps:
    geo_visualizer.create_household_state_figure(time_step, save=True)

geo_visualizer.create_recovery_gif(list(time_steps), load_file_name='2D_buildings_Householdsathome|Timestep:TIME_STEP.png', save_file_name='household_recovery.gif')
```

```python
SELECT_HOUSEHOLD_INDEX = 0 # select a household to print the chat history of

all_households = [hh for c in system.components if hasattr(c, 'households') for hh in c.households]
selected_household = all_households[SELECT_HOUSEHOLD_INDEX] # select a household to print the chat history of
selected_household.print_chat_history()
```
